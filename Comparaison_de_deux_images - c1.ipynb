Dans cet exemple, nous allons essayer de voir comment comparer deux images à l'aide de la bibliothèque skimage.
Pour cela nous allons comparer deux images similaires, puis deux images différentes.

## 1.  importation des bibliothèques

```python
import numpy as np
import pandas as pd
import os
```

```python
import skimage
from skimage import io
from skimage.io import imread
from skimage.io import imshow
from skimage.transform import rotate
import matplotlib.pyplot as plt
from skimage.measure import compare_ssim as ssim
from skimage.transform import resize
```

###  Lire les images

```python
# Lecture d'images dans un tableau
array1 = imread("8867_idx5_x601_y1051_class1.png")
array2 = imread("8867_idx5_x601_y1051_class1.png")
```

# Visualiser les deux images à comparer

```python
fig1, ax1 = plt.subplots(figsize=(2, 2))
fig2, ax2 = plt.subplots(figsize=(2, 2))
print(ax1.imshow(array1))
print(ax2.imshow(array2))
```

####  Voir la dimension de ces images

```python
print(array1.shape)
print(array2.shape)
```

```python
array1flt = array1.astype('float')
array2flt = array2.astype('float')
```

**L'indice de similarité des images va de 0 - 1**

```python
skimage.measure.compare_ssim(array1flt, array2flt,multichannel=True)
```

#### Conclusion : comme on peut voir les deux image ont un indice de similarité égale à 1

# Maintenant avec deux images différentes 

```python
# Lecture d'images dans un tableau
array3 = imread("8867_idx5_x601_y1051_class1.png")
array4 = imread("8867_idx5_x601_y1001_class1.png")
```

```python
fig1, ax1 = plt.subplots(figsize=(2, 2))
fig2, ax2 = plt.subplots(figsize=(2, 2))
print(ax1.imshow(array3))
print(ax2.imshow(array4))
```

```python
print(array3.shape)
print(array4.shape)
```

#### si les images ne sont pas de même dimension, il faut les redimensionner

```python
# array1 = resize(array1, (256, 256))
```

```python
print(array3.shape)
print(array4.shape)
```

```python
array3flt = array3.astype('float')
array4flt = array4.astype('float')
```

**L'indice de similarité des images va de 0 - 1**

```python
skimage.measure.compare_ssim(array3flt, array4flt,multichannel=True)
```

#### on voit que l'indice est différent de 1, ce qui indique une différence entre les deux images même s'ils nous semblent similaires
